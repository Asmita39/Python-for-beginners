#31 Select 10 % of current salary from Employee, 20% of Employee2, 15% of Employee3 and for others 30 % of salary from employee table.

```python
q1=pd.merge(df5,df6,how='inner',left_on='Emp_Id', right_on='Emp_Id')
q1['Emp_Salary_Change_Year'] = pd.to_datetime(q1['Emp_Salary_Change_Year'])
q2=q1.loc[q1.groupby('Emp_Id').Emp_Salary_Change_Year.nlargest(1).index.get_level_values(1)][['Emp_Id','Emp_First_Name','Emp_Middle_Name','Emp_Last_Name','Emp_Salary']].rename(columns={"Emp_Salary": "Current_Salary"})
def Per_Sal (q2):
    if q2['Emp_Id'] == 1:
        calc= q2['Current_Salary']*0.10
    elif q2['Emp_Id'] == 2:
         calc= q2['Current_Salary']*0.20
    elif q2['Emp_Id'] == 3:
         calc = q2['Current_Salary']*0.15
    else:
         calc = q2['Current_Salary']*0.3
    return calc

q2['Calculated_Percentages']=q2.apply(Per_Sal,axis=1)
q2[['Emp_Id','Emp_First_Name','Emp_Last_Name','Current_Salary','Calculated_Percentages']]
```

#33 Select the employee with highest appraisal. (difference of first salary and current salary)

```python
q1=pd.merge(df5,df6,how='inner',left_on='Emp_Id', right_on='Emp_Id')
q1['Emp_Salary_Change_Year'] = pd.to_datetime(q1['Emp_Salary_Change_Year'])
q2=q1.loc[q1.groupby('Emp_Id').Emp_Salary_Change_Year.nlargest(1).index.get_level_values(1)][['Emp_Id','Emp_First_Name','Emp_Middle_Name','Emp_Last_Name','Emp_Salary']].rename(columns={"Emp_Salary": "Current Salary"}).reindex()
q3=q1.loc[q1.groupby('Emp_Id').Emp_Salary_Change_Year.nsmallest(1).index.get_level_values(1)][['Emp_Id','Emp_First_Name','Emp_Middle_Name','Emp_Last_Name','Emp_Salary']].rename(columns={"Emp_Salary": "Initial Salary"}).reindex()
q4=pd.merge(q2,q3,how='inner',left_on=['Emp_Id','Emp_First_Name','Emp_Middle_Name','Emp_Last_Name'], right_on=['Emp_Id','Emp_First_Name','Emp_Middle_Name','Emp_Last_Name'])
q4['Appraisal'] = q4['Current Salary'] - q4['Initial Salary']
q4.sort_values(by='Appraisal',ascending=False).head(1)
```

#35 Display employee name, and average salary with the help of a user defined function.

```python
def Average_Salary (id):
    q1=pd.merge(df5,df6,how='inner',left_on='Emp_Id', right_on='Emp_Id').groupby(['Emp_Id','Emp_First_Name','Emp_Last_Name'])[['Emp_First_Name','Emp_Last_Name','Emp_Salary']].mean().rename(columns={"Emp_Salary": "Average_Salary"})
    return q1.loc[id]
```
